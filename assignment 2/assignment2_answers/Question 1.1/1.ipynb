```python
import pandas as pd
import numpy as np
```

reading the mean_1:

```python
Ds1_m1= pd.read_csv("hwk2_datasets/DS1_m_1.txt", header = None)
Ds1_m1 = Ds1_m1.to_numpy()
Ds1_m1 = Ds1_m1[0][:20]
print(f'the first mean matrix{Ds1_m1}')
```

reading the mean_0:

```python
Ds1_m0= pd.read_csv("hwk2_datasets/DS1_m_0.txt", header = None)
Ds1_m0 = Ds1_m0.to_numpy()
Ds1_m0 = Ds1_m0[0][:20]
print(f'the zero mean matrix{Ds1_m0}')
```

reading the covariance matrix:

```python
Ds1_cov= pd.read_csv("hwk2_datasets/DS1_Cov.txt", header = None)
Ds1_cov
Ds1_cov = Ds1_cov.to_numpy()
Ds1_cov = Ds1_cov[:, :20]
print(f'the covariance matrix{Ds1_cov}')
```

generate data from multivariate Gaussian distribution with Ds1_m1 as mean and Ds1_cov as covariance and label this data to be porisitve( Y = 1)

```python
data_positive = np.random.multivariate_normal(Ds1_m1, Ds1_cov, 2000)

ones = np.ones((2000,1))
data_positive = np.append(data_positive, ones, axis=1)


data_positive_pd = pd.DataFrame(data_positive)
data_positive_pd.head()
```

generate data from multivariate Gaussian distribution with Ds1_m0 as mean and Ds1_cov as covariance and label this data to be negative( Y = 0)

```python
data_negative = np.random.multivariate_normal(Ds1_m0, Ds1_cov, 2000)

ones = 0 * np.ones((2000,1))
data_negative = np.append(data_negative, ones, axis=1) 


data_negative_pd = pd.DataFrame(data_negative)
data_negative_pd.head()

```

pick the train and test and validation data based on the percent in assignment and also pick the data randomly without replacement( by shuffling the data at the first step)

```python
np.random.shuffle(data_positive)
data_positive_train = data_positive[0:1200]
data_positive_test = data_positive[1200:1600]
data_positive_valid = data_positive[1600:2000]
```

```python
np.random.shuffle(data_negative)
data_negative_train = data_negative[0:1200]
data_negative_test = data_negative[1200:1600]
data_negative_valid = data_negative[1600:2000]
```

```python
data_test = np.concatenate((data_positive_test, data_negative_test))
data_train = np.concatenate((data_positive_train, data_negative_train))
data_valid = np.concatenate((data_positive_valid, data_negative_valid))
```

```python
np.random.shuffle(data_test)
np.random.shuffle(data_train)
np.random.shuffle(data_valid)
```

```python
print(f'shape of train data{data_train.shape}')
print(f'shape of test data{data_test.shape}')
print(f'shape of validation data{data_valid.shape}')
```

save the test and train and validation data to the repository

```python
np.savetxt('DS1_train.csv', data_train, delimiter=',')
np.savetxt('DS1_test.csv', data_test, delimiter=',')
np.savetxt('DS1_valid.csv', data_valid, delimiter=',')
```

save the entire datasets to the repository and save it as DS1

```python
data = np.concatenate((data_train, data_test, data_valid))
np.savetxt('DS1.csv', data, delimiter=',')
```
