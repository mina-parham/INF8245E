```python
import numpy as np
#Import Keras
from tensorflow import keras
# Loading the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

```

```python
x_train, y_train, x_valid, y_valid = x_train[:50000], y_train[:50000], x_train[50000:], y_train[50000:]
```

first step: flatten the data

```python
def flatten(arr):
    arr_flat = np.zeros((arr.shape[0], 784))
    for i in range(len(arr)):
        arr_flat[i] = arr[i].flatten()/255
    return arr_flat
```

```python
x_train_flat = flatten(x_train)
x_test_flat = flatten(x_test)
x_valid_flat = flatten(x_valid)
```

```python
def prior(y):
    unique, counts = np.unique(y, return_counts=True)
    return counts/len(y)

def predict(sigma, mean, x, prior_y):
    p = []
    for k in range(10):
        pp = (np.log(prior_y[k]))- np.sum(np.log(sigma[k])) - (1/2) * np.sum(((x - mean[k]) / sigma[k])**2)
        p.append(pp)
                                                                    
    return p,np.argmax([(np.log(prior_y[k]))- np.sum(np.log(sigma[k])) - (1/2) * np.sum(((x - mean[k]) / sigma[k])**2) for k in range(10)])

def mean(x_train_flat, y_train):
    means = np.zeros((10, x_train_flat.shape[1]))
    stdevs = np.zeros((10, x_train_flat.shape[1]))
    
    for k, y_k in enumerate(np.unique(y_train)):
        indices = np.where(y_k == y_train)
        means[k] = np.mean(x_train_flat[indices], axis=0)
        stdevs[k] = np.std(x_train_flat[indices], axis=0) + 0.01
        
    return means, stdevs

def accuracy_df(y, y_hat):
    error = 0
    for i in range(len(y)):
        if y[i] != y_hat[i]:
            error +=1
    return (1 - error/len(y)) * 100

```

### 1.(a)

```python
prior_y = prior(y_train)
means, stdevs = mean(x_train_flat, y_train)

print(f'The prior class probabilities is :\n {prior_y} \n')

for i in range(10):
    print(f'\n mean for class conditional density for class {i} is :  \n \n {means[i]} \n ')
    print(f'\n diagonal covariance matrices for class {i} is : \n \n {stdevs[i]} \n ')



```

### 1.(b)

```python
y_hat = np.zeros((x_test_flat.shape[0]))

for i in range(x_test_flat.shape[0]):
    p, y_hat[i]= predict(stdevs, means, x_valid_flat[i], prior_y)
    
accuracy_ = accuracy_df(y_valid, y_hat)

print(f'best fit accuracy achieved : {accuracy_}')
```
